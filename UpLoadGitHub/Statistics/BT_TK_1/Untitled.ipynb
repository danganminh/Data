```python
import numpy as np
import pandas as pd
import statistics as stt
import os
```

```python
os.getcwd()
```

```python
data = pd.read_excel('data.xlsx')
display(data)
```

```python
Is1 = 70
U1 = data['Unnamed: 0'][0:]
I1 = data['Unnamed: 1'][0:]

Is2 = 50
U2 = data['Unnamed: 2'][0:]
I2 = data['Unnamed: 3'][0:]

Is3 = 40
U3 = data['Unnamed: 4'][0:]
I3 = data['Unnamed: 5'][0:]
```

# Draw Chart of: $ I = f(U) $

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10,6), dpi=300)


# Is = 70
x = np.linspace(min(U1), max(U1), 10000)

plt.scatter(U1,I1, color='red', alpha=0.5)

fit = np.polyfit(U1,I1, 4)
y = fit[0]*x**4 + fit[1]*x**3 + fit[2]*x**2 + fit[3]*x + fit[4]
plt.plot(x,y, color='red', label=r'$I_s = 70(mA)$')

erI1 = stt.stdev(I1) / np.sqrt(len(I1))

plt.fill_between(x, y-erI1, y+erI1, color='red', alpha=.2, linewidth=0)




# Is = 50
x = np.linspace(min(U2), max(U2), 10000)

plt.scatter(U2,I2, color='blue', alpha=0.5)

fit = np.polyfit(U2,I2, 4)
y = fit[0]*x**4 + fit[1]*x**3 + fit[2]*x**2 + fit[3]*x + fit[4]
plt.plot(x,y, color='blue', label=r'$I_s = 50(mA)$')

erI2 = stt.stdev(I2) / np.sqrt(len(I2))

plt.fill_between(x, y-erI2, y+erI2, color='blue', alpha=.2, linewidth=0)




# Is = 40
x = np.linspace(min(U3), max(U3), 10000)

plt.scatter(U3,I3, color='green', alpha=0.5)

fit = np.polyfit(U3,I3, 4)
y = fit[0]*x**4 + fit[1]*x**3 + fit[2]*x**2 + fit[3]*x + fit[4]
plt.plot(x,y, color='green', label=r'$I_s = 40(mA)$')

erI3 = stt.stdev(I3) / np.sqrt(len(I3))

plt.fill_between(x, y-erI3, y+erI3, color='green', alpha=.2, linewidth=0)



# setup figure
plt.ylim([0 , 100])
plt.xlim([0 , 6.5])
plt.title(r'Plot of $I$ = f$(U)$', fontsize = 20, fontname='arial')
L = plt.legend()
plt.setp(L.texts, fontsize=14, family='arial')
plt.xlabel(r'$U$ $(V)$', fontsize=16, fontname='arial')
plt.ylabel(r'$I$ $(mA)$', fontsize=16, fontname='arial')

plt.text(3.5, 92.5, r'$\Delta$ $I$ = $\pm$ ' + str(round(erI1,4)) + r'$(mA)$' , fontsize=13, fontname='arial', color='r')
plt.text(3.5, 92.5-5, r'$\Delta$ $I$ = $\pm$ ' + str(round(erI2,4)) + r'$(mA)$' , fontsize=13, fontname='arial', color='b')
plt.text(3.5, 92.5-5-5, r'$\Delta$ $I$ = $\pm$ ' + str(round(erI3,4)) + r'$(mA)$' , fontsize=13, fontname='arial', color='g')

plt.text(0.5, 20, r'$y = ax^4 + bx^3 + cx^2 + dx + e$', fontsize=15, fontname='arial',)

plt.savefig('bai1.png')
```
