```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

## Gamma Distribution

### gamma function
\begin{align}
\Gamma(\alpha) = \int_0^\infty x^{\alpha - 1} e^{-x} dx
\end{align}
### Gamma distribution

$ f(x) =
  \begin{cases}
    \frac{1}{\Gamma(\alpha)\beta^{\alpha}} x^{\alpha-1} e^{\frac{-x}{\beta}}       & \quad \text{if } x > 0\\
    0  & \quad \text{if } x \leq
  \end{cases}
$

\begin{align}
\int_0^\infty \frac{1}{\Gamma(\alpha)\beta^{\alpha}} x^{\alpha-1} e^{\frac{-x}{\beta}} = 1
\end{align}

```python
# gamma function
x, a, b = sp.symbols('x a b')

gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, sp.oo))

# Gamma distribution
gd = 1/(gf*b**a) * x**(a-1) * sp.exp(-x/b)

display(gf)
display(gd)
```

```python
# Drawplot using numpy and matplotlib

a = 2
b = 2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 10, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='orange', label='G(2,2)')

a = 1
b = 2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 10, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='red', label='G(1,2)')

a = 3
b = 1
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 10, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='blue', label='G(3,1)')

plt.title('Gamma Distribution' + '\n' + r'$G(\alpha \beta)$')
plt.ylabel(r'$f$ $(x)$')
plt.xlabel(r'$x$')
plt.legend()
```

## Chi Square Distribution

\begin{align}
\chi \sim \chi ^2(r)\: \text{if}\: X \sim G(r/2,2)\: \text{with}\: r=1,2,3...
\end{align}

```python
# Draw plot by numpy and matplotlib
b = 2

a = 2/2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 30, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='orange', label='r = 2')

a = 6/2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 30, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='red', label='r = 6')

a = 15/2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 30, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='blue', label='r = 15')

plt.title('Chi Square Distribution' + '\n' + r'$\chi ^2 \sim \chi ^2(r)$')
plt.ylabel(r'$f$ $(x)$')
plt.xlabel(r'$x$')
plt.legend()
```

```python
## Solve value
# exames r = 6

a = 6/2
x = sp.Symbol('x')
gf = sp.integrate(x**(a-1)*sp.exp(-x), (x, 0, np.inf))
x = np.linspace(0, 30, 1000)
f2 = 1/(gf*b**a) * x**(a-1) * np.exp(-x/b)
plt.plot(x, f2, color='deeppink', label='r = 6')

plt.title('Chi Square Distribution' + '\n' + r'$\chi ^2 \sim \chi ^2(r)$')
plt.ylabel(r'$f$ $(x)$')
plt.xlabel(r'$x$')
plt.legend()
plt.ylim([0, 0.15])
```

## Student's Distribution

\begin{align}
T = \frac{X\sqrt n}{Y}
\end{align}

\begin{align}
\text{with}\: X \sim N(0,1), Y \sim \chi ^2(n)
\end{align}

### density function

\begin{align}
f(t) = \frac{\Gamma{((r+1)/2)}}{ \sqrt{\pi r} \Gamma{(r/2)}} \:
\frac{1}{(1+t^2/r)^{(r+1)/2}}
\end{align}

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.Symbol('x')

r = 1
a1 = (r+1)/2
a2 = r/2
gf1 = sp.integrate(x**( a1 -1)*sp.exp(-x), (x, 0, sp.oo))
gf2 = sp.integrate(x**( a2 -1)*sp.exp(-x), (x, 0, sp.oo))
t = np.linspace(-4, 4, 1000)
y = gf1/(np.sqrt(np.pi*r) * gf2) *1/(1+ t**2 / r)**((r+1)/2)
plt.plot(t,y, color='crimson', label='$r=1$')

r = 3
a1 = (r+1)/2
a2 = r/2
gf1 = sp.integrate(x**( a1 -1)*sp.exp(-x), (x, 0, sp.oo))
gf2 = sp.integrate(x**( a2 -1)*sp.exp(-x), (x, 0, sp.oo))
t = np.linspace(-4, 4, 1000)
y = gf1/(np.sqrt(np.pi*r) * gf2) *1/(1+ t**2 / r)**((r+1)/2)
plt.plot(t,y, color='blue', label='$r=3$')

z = np.linspace(-4, 4, 1000)
f = 1/np.sqrt(2*np.pi) * np.exp(-1/2 * z**2)
plt.plot(z,f, color='deepskyblue', label='$N(0,1)$')


plt.title('Density Function Students Distribution')
plt.ylabel('$f$ $(t)$')
plt.xlabel('$t$')
plt.legend()
```
