```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import scipy.special
```

```python
os.getcwd()
```

```python
data = pd.read_excel('data.xlsx')
display(data)
```

```python
X = np.array(data['Số đếm X'][0:14])
N = np.array(data['Tần suất thực nghiệm'][0:14])
print(X)
print(N)
print(sum(N))
```

```python
plt.figure(figsize=(9, 5))

Px = N[0:]/sum(N) # Xs thực nghiệm
meX = sum(X*Px)  # Trung Vị
print(Px)
print(meX)

k = np.linspace(min(X),max(X), 1000)

Y_poi = np.exp(-meX) * (meX)**k / scipy.special.factorial(k)
# Xs lý thuyết (smooth)

# Bar
p1 = plt.bar(X,Px, color='royalblue', width=0.8, label='Xác Suất Thực Nghiệm')


# Plot
p2 = plt.plot(k,Y_poi, color='crimson', linewidth=3.5, label='Xác Suất Poisson')

plt.title("Thống Kê Số Đếm Phông")
plt.ylabel("Xác Suất")
plt.xlabel("Số đếm")
plt.legend()



def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
addlabels(X, Px)
```

```python
X2 = np.array(data['Số đếm X'][0:])
N2 = np.array(data['Tần suất thực nghiệm'][0:])

Px2 = N2[0:]/sum(N2) # Xs thực nghiệm

meX2 = sum(X2*Px2)  # Trung Vị
Y_poi2 = np.exp(-meX2) * (meX2)**X2 / scipy.special.factorial(X2)
# Xs lý thuyết 

df = pd.DataFrame(data)

df['Xác suất thực nghiệm'] = Px2
df['Xác suất Poisson'] = Y_poi2


display(df)
```
