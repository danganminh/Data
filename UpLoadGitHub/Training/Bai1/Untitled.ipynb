```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
os.getcwd()
```

```python
dataset = pd.read_excel("data1-2.xlsx")
pd.options.display.max_columns = None
display(dataset)
```

```python
I = dataset["I"].values.tolist()
B = dataset["B"].values.tolist()

x = np.array(I)
y = np.array(B)
n = len(x)

plt.style.use('default')

# Draw Scatter
s1 = plt.scatter(x,y, color='b', label="Real Data")

# Cal Error Number
S2x = sum((x-np.mean(x))**2)/(n-1)
Sx = np.sqrt(S2x)
errx = Sx/(np.sqrt(n))
print('errx = ' + str(round(errx,4)))

S2y = sum((y-np.mean(y))**2)/(n-1)
Sy = np.sqrt(S2y)
erry = Sy/(np.sqrt(n))
print('erry = ' + str(round(erry,4)))

plt.errorbar(x,y, yerr=erry, color='black', linestyle='none', capsize=5)


# Poly fit
Fit = np.polyfit(x,y, 1)
x_smt = np.linspace(np.min(x), np.max(x), 100)
yfit = Fit[0]*x_smt + Fit[1]
print('linear y = ' + str(round(Fit[0],4)) + 'x' + str(round(Fit[1],4)) )
p1 = plt.plot(x_smt,yfit, color='r', label="Interpolation")

# regression linear
y_in = Fit[0]*x + Fit[1]
R2 = 1 - (sum((y-y_in)**2)) / (sum((y-np.mean(y))**2))
print("R^2 = " + str(round(R2,4)) )


plt.title("Plot of $B = f(I)$")
plt.xlabel("I $(A)$")
plt.ylabel("B $(mT)$")
plt.legend(loc="upper left")
plt.text(1.5,0.4, "y = " + str(round(Fit[0],4)) + "x" + str(round(Fit[1],4)) )
plt.text(1.5,0.3, "$R^2 = $" + str(round(R2,4)))

plt.text(0,1,"$\Delta B = $" + str(round(erry,4)) )
plt.text(0,0.9,"$\Delta I = $" + str(round(errx,4)) )
```

```python
dataset2 = pd.read_excel("data2.xlsx")
pd.options.display.max_columns = None
display(dataset2)
```

```python
z = dataset2["z"].values.tolist()
B = dataset2["B"].values.tolist()

x = np.array(z)
y = np.array(B)
n = len(y)

plt.style.use('default')

# Draw Scatter
s = plt.scatter(x,y, color='b', label='Real Data')

# Cal Error Number
Sy2 = sum((y-np.mean(y))**2) / (n-1)
Sy = np.sqrt(Sy2)
erry = Sy/(np.sqrt(n))
print("erry = " + str(round(erry,4)) )
plt.errorbar(x,y, yerr=erry, color='black', linestyle='none', capsize=5)

# Poly Fit
fit = np.polyfit(x,y, 2)
xsm = np.linspace(min(x), max(x), 100)
yfit = fit[0]*xsm**2 + fit[1]*xsm + fit[2]
p = plt.plot(xsm,yfit, color='r', label='Interpolation')
print("Interpolation y = " + str(round(fit[0],4)) + 'x^2 + ' + str(round(fit[1],4)) + 'x' + str(round(fit[0],4)))

# regression linear
y_in = fit[0]*x**2 + fit[1]*x + fit[2]
R2 = 1 - (sum((y-y_in)**2)) / (sum((y-np.mean(y))**2))
print("R^2 = " + str(round(R2,4)) )

plt.legend(loc='best')
plt.text(-7,0.9, "y = " + str(round(fit[0],4)) + "$x^2$ + " + str(round(fit[1],4)) + 'x + ' + str(round(fit[2],4)) )
plt.text(-7,0.85, "$R^2 = $" + str(round(R2,4)) )
plt.text(8,1.2, "$\Delta B = $" + str(round(erry,4)) )
plt.title("Plot of B = f(z)")
plt.xlabel("z $(cm)$")
plt.ylabel("B $(mT)$")



```
