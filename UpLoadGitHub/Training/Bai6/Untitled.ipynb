```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
data = pd.read_excel('data.xlsx')
display(data)
```

```python
plt.figure(figsize=(8, 5), dpi=(200))

x = np.array(data['Phi'][0:])
y = np.array(data['I'][0:])

xsm = np.linspace(min(x)-5, max(x)+5, 10000)

# Scatter Chart
c1 = plt.scatter(x,y)

# Poly Fit 2
f2 = np.polyfit(x,y,2)
y2 = f2[0]*x**2 + f2[1]*x + f2[2]
R2_2 = 1 - (sum((y-y2)**2)) / (sum((y-np.mean(y))**2))
print(R2_2)
y2sm = f2[0]*xsm**2 + f2[1]*xsm + f2[2]


# Poly Fit 4
f4 = np.polyfit(x,y,4)
y4 = f4[0]*x**4 + f4[1]*x**3 + f4[2]*x**2 + f4[3]*x + f4[4]
R2_4 = 1 - (sum((y-y4)**2)) / (sum((y-np.mean(y))**2))
print(R2_4)
y4sm = f4[0]*xsm**4 + f4[1]*xsm**3 + f4[2]*xsm**2 + f4[3]*xsm + f4[4]


# Poly Fit 6
f6 = np.polyfit(x,y, 6)
y6 = f6[0]*x**6 + f6[1]*x**5 + f6[2]*x**4 + f6[3]*x**3 + f6[4]*x**2 + f6[5]*x + f6[6]
R2_6 = 1 - (sum((y-y6)**2)) / (sum((y-np.mean(y))**2))
print(R2_6)
y6sm = f6[0]*xsm**6 + f6[1]*xsm**5 + f6[2]*xsm**4 + f6[3]*xsm**3 + f6[4]*xsm**2 + f6[5]*xsm + f6[6]


# Plot Chart
p1 = plt.plot(xsm, y2sm, color='red', linestyle=':',label='PolyFit 2')
p2 = plt.plot(xsm, y4sm, color='darkblue', linestyle='-.',label='PolyFit 4')
p3 = plt.plot(xsm, y6sm, color='green', label='PolyFit 6')

# Plot title and label
plt.title(r'Chart of I = f $(\theta)$')
plt.xlabel(r'$\theta ^o$')
plt.ylabel('I $(A)$')
plt.legend()

# Error Number
n = len(y)
S2y = sum((y-np.mean(y))**2)/(n-1)
Sy = np.sqrt(S2y)
ery = Sy/(np.sqrt(n))
plt.errorbar(x,y, yerr=ery, 
            color='black', linestyle='none', capsize=5)

# TEXT
plt.text(155,0.65, r'$\Delta I = \pm$' + str(round(ery,4)) + '$(A)$')
```
