```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
data = pd.read_excel('data.xlsx')
pd.options.display.max_columns = None
display(data)
```

```python
oc = (data["Occupation"][0:14])
ma = (data["Males"][0:14])
fe = (data["Females"][0:14])
```

```python
plt.style.use('_mpl-gallery')
plt.figure(figsize=(5, 4.5))

h1 = plt.bar(oc, ma, color="dodgerblue")
plt.xticks(rotation=90)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

addlabels(oc, ma)
plt.title("Males Employees")
plt.ylabel("Employees")
plt.grid(True)
```

```python
plt.style.use('_mpl-gallery')
plt.figure(figsize=(5, 4.5))

h2 = plt.bar(oc, fe, color="lightpink")
plt.xticks(rotation=90)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

addlabels(oc, fe)
plt.title("Females Employees")
plt.ylabel("Employees")
plt.grid(True)
```

```python
tot = np.array([sum(ma), sum(fe)])
plt.style.use('_mpl-gallery-nogrid')

pielabels = ["Males", "Females"]
piecolor = ["dodgerblue", "deeppink"]
pieexplode = [0.05, 0]



# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n{:d}".format(pct, absolute)
                    
plt.figure(figsize=(5, 5))
                                      
p1 = plt.pie(tot, labels = pielabels, 
             colors= piecolor, shadow = True, 
             explode = pieexplode,
             autopct = lambda pct: func(pct, tot))
```
